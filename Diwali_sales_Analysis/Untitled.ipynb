## Exploratory Data Analysis

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers: #7842 dekhauna
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

```python
ap=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ap.containers: #7842 dekhauna
    ap.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum()
```

```python
sns.barplot(x="Age Group",y="Amount",data=sales_age)
```

# State

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=sales_state,x='State',y="Orders")
```

```python
sales_Amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sales_Amount
```

```python
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=sales_Amount,x='State',y="Amount")
```
